```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Startups_Expense.csv')
```

```python
df.head()
```

```python
df.head(10)
```

```python
# so we observed state data is into 3 main values - NY, FL, CA
# so out of those 3, we will ignore  1 - to come with dummy variable trap
```

```python
df1 = pd.get_dummies(df['State'],drop_first=True)
df1
```

```python
# we will do concatenation in order to append this  false/true as column and assigning as new dataset
df = pd.concat([df1,df],axis=1)
df
```

```python
# now we want to drop manin column - states ; with encoding NY=1, FL=0 
df.drop('State',axis=1,inplace=True)
```

```python
df.head()
```

```python
# splitting data into x and y = input and output data
# x is taking all rows, exclusing last column which is profit
# y is taking all rows but excluding 5 columns = profit is output
X=df.iloc[:,:-1]
y=df.iloc[:,[5]]
```

```python
# 4 paramters are needed to train and test split; here test size is 25%
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
# apply linear regression with input and output training data
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
#LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

```python
# find out y_predict based on x_test
y_pred=regression.predict(X_test)
```

```python
# find accuracy r- by comparing predicted value and test o/p value
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```
